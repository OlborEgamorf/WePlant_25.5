```python
import pandas as pd
```

```python
df = pd.read_csv("../data/A_to_Z_Flowers_cleaned.csv")
df.head()
```

```python
def calculer_score_Sun(sun_value):
    score = 0
    if "Full sun" in sun_value:
        score += 1
    if "Partial sun" in sun_value:
        score += 0.5
    if "Shade" in sun_value:
        score += 0
    return score

df["SunNeeds_Score"] = df["SunNeeds"].apply(calculer_score_Sun)

df[["SunNeeds", "SunNeeds_Score"]].head()
```

```python
def calculer_score_Water(water_value):
    score = 0
    if "avarage" in water_value:
        score += 1
    if "low" in water_value:
        score += 0
    return score

df["WaterNeeds_Score"] = df["WaterNeeds"].apply(calculer_score_Water)

df[["WaterNeeds", "WaterNeeds_Score"]].head()
```

```python
def calculer_score_Maintenance(maintenance_value):
    score = 0
    if "hard" in maintenance_value:
        score += 1
    if "low" in maintenance_value:
        score += 0.5
    if "medium" in maintenance_value:
        score += 0
    return score

df["Maintenance_score"] = df["Maintenance"].apply(calculer_score_Maintenance)

df[["Maintenance", "Maintenance_score"]].head()
```
