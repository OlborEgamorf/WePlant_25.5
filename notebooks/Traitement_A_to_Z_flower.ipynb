```python
import pandas as pd
```

```python
df = pd.read_csv("../data/A_to_Z_Flowers_cleaned.csv")
df.head()
```

```python
# Suppression des doublons sur toutes les colonnes
df_cleaned = df.drop_duplicates()
print("Nombre d'enregistrements après suppression des doublons :", len(df_cleaned))
```

```python
def analyser_besoins_soleil(sun_value):
    score = 0
    sun_fr = []
    
    if isinstance(sun_value, str):
        if "Full sun" in sun_value:
            score += 1
            sun_fr.append("Plein soleil")
        if "Partial sun" in sun_value:
            score += 0.5
            sun_fr.append("Mi-ombre")
        if "Shade" in sun_value:
            score += 0
            sun_fr.append("Ombre")
    
    return score, ", ".join(sun_fr)

# Appliquer la fonction
df["SunNeeds_Score"], df["SunNeeds"] = zip(*df["SunNeeds"].apply(analyser_besoins_soleil))

# Afficher les résultats
print(df[["SunNeeds", "SunNeeds_Score"]].head())
```

```python

def calculer_score_Water(water_value):
    score = 0    
    water = []
    if isinstance(water_value, str):
        water_value = water_value.lower()
        
        if "average" in water_value or "avarage" in water_value:
            score += 1
            water.append("Moyenne")
        if "low" in water_value:
            score += 0
            water.append("Faible")
        if "high" in water_value:
            score += 2
            water.append("Elevée")
        
        # Gérer les cas spéciaux
        if "low to average" in water_value:
            score += 0.5
            water.append("Faible à moyenne")
        if "average to high" in water_value:
            score += 1.5    
            water.append("Moyenne à élevée")   
    
    return score, ", ".join(water)


df["WaterNeeds_Score"], df["WaterNeeds"] = zip(*df["WaterNeeds"].apply(calculer_score_Water))
df[["WaterNeeds", "WaterNeeds_Score"]].head()
```

```python
def calculer_score_Maintenance(maintenance_value):
    score = 0
    maintenance = []
    if isinstance(maintenance_value, str):
        maintenance_value = maintenance_value.lower()
        if "hard" in maintenance_value:
            score += 1
            maintenance.append("Difficile")
        if "low" in maintenance_value:
            score += 0
            maintenance.append("Faible")
        if "medium" in maintenance_value:
            score += 0.5
            maintenance.append("Moyen")
    return score, ", ".join(maintenance)

df["Maintenance_score"], df["Maintenance"] = zip(*df["Maintenance"].apply(calculer_score_Maintenance))
df[["Maintenance", "Maintenance_score"]].head()
```

```python
def saison(saison_value):
    saisons = []
    if isinstance(saison_value, str):  # Check if the value is a string
        if "Spring" in saison_value:
            saisons.append("Printemps")
        if "Autumn" in saison_value:
            saisons.append("Automne")
        if "Summer" in saison_value:
            saisons.append("Été")
        if "Winter" in saison_value:
            saisons.append("Hiver")
    return ", ".join(saisons)

# Apply the function
df["saison"] = df["BloomsIn"].apply(saison)

# Display results
print(df[["BloomsIn", "saison"]].head())
```

```python
# Fonction pour catégoriser ou traduire les types de plantes
def categorize_plant_types(plant_value):
    categories = []
    if isinstance(plant_value, str):  # Vérifier que la valeur est une chaîne
        if "Annuals" in plant_value:
            categories.append("Annuelles")
        if "Perennials" in plant_value:
            categories.append("Vivaces")
        if "Bulbs" in plant_value:
            categories.append("Bulbes")
        if "Shrubs" in plant_value:
            categories.append("Arbustes")
        if "Trees" in plant_value:
            categories.append("Arbres")
        if "Climbers" in plant_value:
            categories.append("Grimpantes")
        if "Herbs" in plant_value:
            categories.append("Herbes")
        if "Cactus - Succulents" in plant_value:
            categories.append("Cactus - Succulentes")
        if "Roses" in plant_value:
            categories.append("Roses")
    return ", ".join(categories)

# Appliquer la fonction à la colonne PlantType
df["plant_categories"] = df["PlantType"].apply(categorize_plant_types)

# Afficher les résultats
print(df[["PlantType", "plant_categories"]].head())
```

```python
import re
import pandas as pd

def extract_height(height_text):
    if not isinstance(height_text, str):
        return None, None  # Gérer les valeurs non-string

    # Expressions régulières pour trouver les hauteurs en différentes unités
    cm_pattern = re.compile(r'(\d+(?:,\d+)?(?:\.\d+)?) cm', re.IGNORECASE)
    m_pattern = re.compile(r'(\d+(?:,\d+)?(?:\.\d+)?) m', re.IGNORECASE)
    ft_pattern = re.compile(r'(\d+(?:,\d+)?(?:\.\d+)?) ft', re.IGNORECASE)
    inch_pattern = re.compile(r'(\d+(?:,\d+)?(?:\.\d+)?) inch(?:es)?', re.IGNORECASE)

    def convert_to_cm(value, unit):
        value = float(str(value).replace(',', '.'))
        if unit == 'm':
            return value * 100
        elif unit == 'ft':
            return value * 30.48
        elif unit == 'inch':
            return value * 2.54
        else:
            return value  # Assume cm

    # Trouver toutes les correspondances et leurs unités
    matches = []
    matches.extend((m.group(1), 'cm') for m in cm_pattern.finditer(height_text))
    matches.extend((m.group(1), 'm') for m in m_pattern.finditer(height_text))
    matches.extend((m.group(1), 'ft') for m in ft_pattern.finditer(height_text))
    matches.extend((m.group(1), 'inch') for m in inch_pattern.finditer(height_text))

    if not matches:
        return None, None  # Aucune hauteur trouvée

    # Convertir toutes les hauteurs en cm
    heights_cm = [convert_to_cm(value, unit) for value, unit in matches]

    # Trouver la hauteur minimale et maximale
    min_height = min(heights_cm)
    max_height = max(heights_cm)

    return min_height, max_height

# Appliquer la fonction à la colonne 'Height'
df['min_height_cm'], df['max_height_cm'] = zip(*df['Height'].apply(extract_height))

# Afficher les résultats
print(df[['Height', 'min_height_cm', 'max_height_cm']].head())
```

```python
import re
import pandas as pd

def extract_ph(soil_text):
    if not isinstance(soil_text, str):
        return None, None  # Gérer les valeurs non-string

    # Expression régulière pour trouver les plages de pH
    ph_pattern = re.compile(r'pH\s*(?:between|of|from)?\s*([4-9](?:\.\d+)?)\s*(?:to|-)?\s*([4-9](?:\.\d+)?)?', re.IGNORECASE)
    match = ph_pattern.search(soil_text)

    if match:
        min_ph = (match.group(1))
        max_ph = (match.group(2)) if match.group(2) else min_ph  # Si une seule valeur est trouvée

        return min_ph, max_ph
    else:
        return None, None

df['min_ph'], df['max_ph'] = zip(*df['SoilNeeds'].apply(extract_ph))

# Afficher les résultats
print(df[['SoilNeeds', 'min_ph', 'max_ph']].head())
```
